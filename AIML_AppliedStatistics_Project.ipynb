```python
#1.A
    #Refer above table and find the joint probability of the people who planned to purchase and actually placed an order.

#Solution:
#---------------------------------------------------------
# Let A be an event - "Planned to Purchase Product A"
# and B be an event - "Actually Placed Order for Product A"
    
# Total Number of Observations = 2000

# Number of People who planned to purchase and placed an order A = 400
    
# P(A and B) = (Number of People who planned to Purchase and actually placed an order for Product A) / (Total Number of People)
   
# Answer:
#P(A and B) = 400/2000

print ("P(A and B) = 400/2000")
```

```python
#1.B
    #Refer to the above table and find the joint probability of the people who planned to purchase and actually placed an order, given that 
    #people planned to purchase.
    
#Solution:
#---------------------------------------------------------
# This is a Conitional Probability

# Let A be an event - "Planned to Purchase Product A"
# and B be an event - "Actually Placed Order for Product A"
    
# Total Number of Observations = 2000

# Total Number of People who planned to Purchase Product A (Let it be m) = 500
# Number of People who planned to purchase and placed an order A (Let it be n)= 400
    
# By using following formula:
#     P(B | A) = P(B and A)/P(A)
    
# Given, P(A)=500/2000
#         P(A and B) = 400/2000
    
#     P(B | A) = (400/2000)/(500/2000) = 400/500

# Answer:
# P(B | A)= 4/5

print('P(B | A)= 4/5')
```

```python
#2
    # An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the 
    # failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following 
    # questions.

    # 2.A. Probability that none of the items are defective? [1 Mark]
    # 2.B. Probability that exactly one of the items is defective? [1 Mark]
    # 2.C. Probability that two or fewer of the items are defective? [1 Mark]
    # 2.D. Probability that three or more of the items are defective ? [1 Mark]

    
#Solution:
#---------------------------------------------------------

#Firm : Electrical Manufacturing Company
    
#Failure Rate of the manufactured item is 5% : P(Failure)=0.05

#Random Sample of 10 items is selectd : n=10

n=10 # sample Size
p=0.05 #Probability that manufactured item is defective

#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import scipy.stats as stats
from scipy.stats import binom

X=np.arange(0,11) # Provides all the possible number of Defective items.
#which can range from 0(none are defective) to 10 (All are defective item)

binomial_prob = binom.pmf(k=X,n=n,p=p)

print("Binomial Probability for each random variable is :")
print(binomial_prob)
print()
print("*********Answers*********")

print("2.A> Probability that none of the items are defective = ",binomial_prob[0])
print("2.B> Probability that exactly one of the items is defective = ",binomial_prob[1])
print("2.C> Pobability that two or fewer of the items are defective = ",binom.cdf(k=2,n=n,p=p))
print("2.D> Probability that three or more of the items are defective = ",1-binom.cdf(k=2,n=n,p=p))
```

```python
#3.
    # A car salesman sells on an average 3 cars per week.
    # 3.A. What is Probability that in a given week he will sell some cars?
    # 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?
    # 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

#Solution:
#---------------------------------------------------------
#Observations
# 1. Average sell per week = 3
# 2. Number of trials not mentioned
# 3. Sample Size is not mentioned
# 4. It follows discrete distribution as Random variable is discrete

# As observed above, we can use Poisson distribution Probability mass function (pmf)
# P(X=x) = (e^-λ * λ^x ) /x!

#What is Probability that in a given week he will sell some cars? 

from scipy.stats import poisson

Rv=np.arange(0,20) # Random Variable 
# Even though not required, if I have a array of probabilities, it will help in calculation

poisson_dist= poisson.pmf(k=Rv,mu=3)

print("2.A> Probability that in a given week he will sell some cars = ",1-poisson_dist[0])
print("2.A> Probability that in a given week he will sell 2 or more but less than 5 cars = ",
      poisson_dist[2]+poisson_dist[3]+poisson_dist[4])

poisson_cdf= poisson.cdf(k=Rv,mu=3)

plt.plot(Rv,poisson_cdf)
plt.title("CDF vs Cars Sold Per Week ")
plt.xlabel("Cars Sold Per Week")
plt.ylabel("Cumulative Probability")
plt.show()
```
